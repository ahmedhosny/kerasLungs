```r
drops <- c("dataset","patient","pathToData","pathToMask","stackMin","stackMax","sizeX","sizeY","sizeZ",
          "orgSpacX","orgSpacY","orgSpacZ","voxelCountList","chosenMaskIdx","bboxX","bboxY","bboxZ","comX",
          "comY","comZ","fail" ,"dense2_0","dense2_1","logit_0","logit_1")
retest <- retest[ , !(names(retest) %in% drops)]
retest
```

```r
counter <- 0
#
result <- c()
#
# loop through 256 features
for (i in 2:257){
    sika <- cbind(test[,i], retest[,i])
    print ( sika[1,1] ) 
    print ( sika[1,2] ) 
    icca <- icc(sika, model="twoway", type = "a")$value
    print (icca)
    result <- c(result,icca)
    counter = counter +1
}
print (counter)
```
